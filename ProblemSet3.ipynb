```python
# Step 1
import pandas as pd
import datetime

# step 2
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data"

# step3
data = pd.read_table(url, sep='\s+')

DateTime = data['Yr'].astype(str)+'/'+data['Mo'].astype(str)+'/'+data['Mo'].astype(str)
data.insert(0, 'Date', DateTime)
data['Date'] =  pd.to_datetime(data['Date'])

del data["Yr"]
del data["Mo"]
del data["Dy"]

data.head()
```

```python
#Step4

def Reset(t):
    
    if t.year > 2000:
        year = t.year-100
        
    else:
        year = t.year
        
    return datetime.date(year, t.month, t.day)

data['Date'] = data['Date'].apply(Reset)

data.head()
```

## Question 6
Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

## Solution:

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

MaDi = pd.read_csv("us-marriages-divorces-1867-2014.csv")

lineplot = plt.gca()

MaDi.plot(x='Year',y='Marriages', color='green', ax=lineplot)
MaDi.plot(x='Year',y='Divorces', color='red', ax=lineplot)

plt.xlabel("Year")
plt.ylabel("Per Capita")

lineplot.set_title('Marriages and Divorces per Capita in U.S. [1867-2014]')

plt.show()
```

## Question 7
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

## Solution

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

Compare = pd.read_csv('us-marriages-divorces-1867-2014.csv')

NewMaDi = Compare.loc[(Compare['Year'] == 1900)|(Compare['Year'] == 1950)|(Compare['Year'] == 2000)]

NewMaDi.plot(x="Year", y=["Marriages_per_1000", "Divorces_per_1000"], kind="bar")

plt.xlabel('Year')
plt.ylabel('Per Capita')

plt.title('Number of Marriages and Divorces per capita in U.S. [1900, 1950, & 2000]')

plt.show()
```

## Question 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

## Solution

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

Actors = pd.read_csv('actor_kill_counts.csv')

Actors = Actors.sort_values(by=['Count'])

plt.barh(Actors["Actor"], Actors["Count"], color='red')

plt.xlabel('Kill Count')
plt.ylabel('Deadliest Actors')

plt.title('Top Deadliest Actors in Hollywood')

plt.show()
```

## Question 9
Create a pie chart showing the fraction of all Roman Emperors that were assassinated. Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

## Solution:

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

RomansEmperor = pd.read_csv("roman-emperor-reigns.csv")

Death =RomansEmperor.loc[(RomansEmperor['Cause_of_Death']=='Assassinated')]

PieLabel = Death['Emperor']

plot = Death.plot.pie(y='Length_of_Reign',figsize=(10, 10), labels = PieLabel, autopct='%1.1f%%',textprops={'fontsize': 12})

plt.axis('equal')

plt.title('Cause of death of Roman Emperors')

plt.show()
```
